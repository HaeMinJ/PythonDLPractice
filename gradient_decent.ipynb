```python
# z = x^2+y^2

import numpy as np
import matplotlib.pyplot as plt

cycle = 30
eta = 0.1 #Learning rate
x = [0]*(cycle+1)
y = [0]*(cycle+1)
z = [0]*(cycle+1)
x[0] = 3
y[0] = 2
for i in range(cycle):
    gradX = 2*x[i]
    gradY = 2*y[i]
    delX = -eta * gradX
    delY = -eta * gradY
    x[i+1] = x[i] + delX
    y[i+1] = y[i] + delY
    z[i] = x[i]*x[i] + y[i]*y[i]

plt.plot(z)
plt.show()
```
